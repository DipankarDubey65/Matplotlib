```python
import matplotlib.pyplot as plt
import pandas as pd
x =['day1','day2','day3','day4','day5']
y =[300,420,250,230,400]

plt.plot(x,y,marker = '*',color ='red')
plt.show()
```

```python

plt.plot(x,y,marker ='o', ls=':',color ='red')
plt.show()
```

```python
x =['day1','day2','day3','day4','day5']
y =[300,420,250,230,400]
y2 =[500,450,300,250,320]

plt.plot(x,y,marker ='o',color ='red',label = 'week1')
plt.plot(x,y2,marker = '*',color='green',label ='week2')
plt.legend()
plt.show()
```
