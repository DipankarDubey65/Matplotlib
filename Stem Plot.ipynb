```python
import matplotlib.pyplot as plt
x = [10,40,20,40,20,40,60,70,50,40]
plt.stem(x)
plt.show()
```

```python
plt.stem(x,linefmt = '--', markerfmt ='D')
plt.show()
```

```python
plt.stem(x,linefmt = '--', markerfmt ='D',orientation ='horizontal')
plt.show()
```
