```python
import matplotlib.pyplot as plt
import pandas as pd
brands =['Onepluse','Apple','Nokia','Redmi','Sumsung']
x =[22,35,30,3,10]

plt.pie(x)
plt.show()
```

```python
plt.pie(x, labels= brands)
plt.show()
```

```python

c =['red','yellow','blue','pink','orange']
exp =[0,0,0,0,0.2]
plt.pie(x, labels = brands, colors = c, explode =exp ,shadow =True)
plt.show()
```

```python

c =['red','yellow','blue','pink','orange']
brands =['Onepluse','Apple','Nokia','Redmi','Sumsung']
exp =[0,0,0,0,0.2]
x =[22,35,30,3,10]
plt.pie(x, labels = brands, colors = c, explode =exp ,shadow =True,autopct = '%.2f', startangle = 90)
plt.savefig('Pies.png')
plt.show()
```
