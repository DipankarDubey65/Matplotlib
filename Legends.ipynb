```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [45,67,33,67,12]
y2 =[41,60,13,66,13]

plt.plot(x,y1 , label = 'male')
plt.plot(x,y2, label = 'female')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = [10,3])
plt.plot(x,y1 , label = 'male')
plt.plot(x,y2, label = 'female')
plt.legend()
plt.show()
```

```python
y3 = [11,30,55,60,19]

plt.figure(figsize = [10,5])
plt.plot(x,y1 , label = 'male')
plt.plot(x,y2, label = 'female')
plt.plot(x,y3, label = 'keeds')
plt.legend(['a1','a2','a3']) # label rename
plt.show()
```

```python
y4 =[11,40,45,90,59]
plt.figure(figsize = [8,5])
plt.plot(x,y1, label ='Male')
plt.plot(x,y2, label = 'Female')
plt.plot(x,y3, label = 'Keeds')
plt.plot(x,y4, label = 'Blanks')

plt.legend(bbox_to_anchor = (0.8, 1.2), ncol = 2)

plt.show()

```
